## A function for making a unique working directory

```python
def mkWorkDir(basedir):
    import os, uuid
    uniqueID=str(uuid.uuid4())
    workDir=basedir+'ipengine-'+uniqueID
    os.mkdir(workDir)
    return workDir
dview.push(dict(mkWorkDir=mkWorkDir));
```

## A function for writing a LAMMPS input script

```python
def lmpscript(
        workDir,
        lmpName,
        potential,
        dx,
        dy,
        file0,
        file1
    ):
    import os
    
    os.chdir(workDir)
    
    f = open(lmpName, 'w')
    f.write('units metal\n')
    f.write('atom_style atomic\n')
    f.write('lattice fcc 3.52 orient x 0.0 -1.0 1.0 orient y 2.0 -1.0 -1.0 orient z 1.0 1.0 1.0 origin 0.0001 0.0001 0.0001 spacing 1.41421356237 2.44948974278 1.73205080757\n')
    f.write('boundary p p m\n')
    f.write('region box block 0 2.0 0 1.0 0 6.0\n')
    f.write('region r1 block 0 2.0 0 1.0 0 3.0\n')
    f.write('region r2 block 0 2.0 0 1.0 3.0 6.0\n')
    f.write('create_box 2 box\n')
    f.write('create_atoms 1 region box\n')
    f.write('pair_style eam/alloy\n')
    f.write('pair_coeff * * %s Ni Al\n' % potential)
    f.write('neighbor 1.0 bin\n')
    f.write('neigh_modify delay 0\n')
    f.write('thermo_style custom step temp press pe ke etotal vol lx ly lz pxx pyy pzz pxy pxz pyz\n')
    f.write('group g2 region r2\n')
    f.write('variable pe equal "pe"\n')
    f.write('variable pe_per_atom equal pe/atoms\n')
    f.write('variable l1 equal "lx"\n')
    f.write('variable l2 equal "ly"\n')
    f.write('variable l3 equal "lz"\n')
    f.write('variable nAtoms equal count(all)\n')
    f.write('fix print all print 1 "${nAtoms} ${l1} ${l2} ${l3} ${pe} ${pe_per_atom}" file %s\n' % file0 )
    f.write('run 1\n')
    f.write('unfix print\n')
    f.write('displace_atoms  g2 move %e %e 0.0\n' % (dx,dy))
    f.write('fix relaxz all setforce 0.0 0.0 NULL\n')
    f.write('minimize 0.0 1.0e-12 1000 100000\n')
    f.write('min_modify dmax 0.0001\n')
    f.write('fix print all print 1 "${nAtoms} ${l1} ${l2} ${l3} ${pe} ${pe_per_atom}" file %s\n' % file1 )
    f.write('run 1\n')
    f.write('unfix print\n')
    f.write('f.close()\n')
dview.push(dict(lmpscript=lmpscript));
```

## A function for executing a LAMMPS input script

```python
def runlmp(workDir,lmpExe,lmpName,cpus):
    import os, subprocess
    os.chdir(workDir)
    string='mpirun -np '+str(cpus)+' '+lmpExe+' < '+lmpName
    subprocess.call([string],shell=True)
dview.push(dict(runlmp=runlmp));
```

## A function for saving data to the HDF5 file

```python
def recordResult(
        resultsFile,
        dx,
        dy,
        lx,
        ly,
        e0,
        e1
    ):
    import tables
    h5file=tables.openFile(resultsFile, mode='r+')
    table=h5file.root.simulation.results
    simulation=table.row
    
    simulation['dx']=dx
    simulation['dy']=dy
    simulation['lx']=lx
    simulation['ly']=ly
    simulation['e0']=e0
    simulation['e1']=e1
    
    simulation.append()
    h5file.close()
dview.push(dict(recordResult=recordResult));
hview.push(dict(recordResult=recordResult)); 
```

## A parallel function to define the simulation *workflow*

```python
@lview.parallel()
def simulation(
        basedir,
        resultsFile,
        cpus,
        lmpExe,
        potential,
        dx,
        dy
    ):
    import numpy
    
    # Make a working directory
    workDir=mkWorkDir(basedir)
    
    # Write LAMMPS script
    lmpName="in.gsfe"
    file0="data.0.txt"
    file1="data.1.txt"
    lmpscript(workDir,lmpName,potential,dx,dy,file0,file1)
    
    # Run LAMMPS
    runlmp(workDir,lmpExe,lmpName,cpus)
    
    # Read results
    data0 = numpy.loadtxt(file0)
    data1 = numpy.loadtxt(file1)
    
    lx=data0[1]
    ly=data0[2]
    e0=data0[4]
    e1=data1[4]
    
    # Store results
    hview.apply_sync(
        recordResult,
        resultsFile,
        dx,
        dy,
        lx,
        ly,
        e0,
        e1
    )
```

## Prepare *workflow* input parameters

```python
INdx=[]
INdy=[]
sims=0

for dx in numpy.linspace(0,0.5,101):
    for dy in numpy.linspace(0,0.5,101):
        sims+=1
        INdx.append(float(dx))
        INdy.append(float(dy))
print sims

INbasedir     = [ str(basedir) for i in xrange(sims) ]
INresultsFile = [ str(resultsFile) for i in xrange(sims) ]
INcpus        = [ cpus for i in xrange(sims) ]
INlmpExe      = [ str(lmpExe) for i in xrange(sims) ]
INpotential   = [ str(potential) for i in xrange(sims) ]
```

## Map *workflow* to IPython Cluster

```python
simulation.map(
    INbasedir,
    INresultsFile,
    INcpus,
    INlmpExe,
    INpotential,
    INdx,
    INdy               
);
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm

a=numpy.array( [ 
        [row['dx'],row['dy'],(row['e1']-row['e0'])/(row['lx']*row['ly'])*16021.7657] 
        for row in results.iterrows()
     ] )
points=a[numpy.lexsort((a[:,2],a[:,1],a[:,0]))]
Xrange=sorted(set(points[:,0]))
Yrange=sorted(set(points[:,1]))
X,Y=numpy.meshgrid(Xrange,Yrange)
Z=griddata(points[:,0],points[:,1],points[:,2],X,Y)

fig=plt.figure(figsize(12,6))

matplotlib.rcParams.update({'font.family' : 'serif','font.size': 16})

ax=fig.gca(projection='3d')
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.coolwarm,linewidth=0,antialiased=False)
ax.view_init(70,15)

ax.set_xlabel("\n$[0\,\overline{1}\,1]$ shift $\Delta$ (fraction)",fontsize=16,linespacing=1.5)
ax.set_ylabel("\n$[2\,\overline{1}\,\overline{1}]$ shift $\Delta$ (fraction)",fontsize=16,linespacing=2.0)
ax.set_zlabel("\nFault energy ($mJ/m^2$)",fontsize=16,linespacing=1.5)

plt.show()



fig,ax=subplots()

p = ax.pcolor(Y, X, Z, cmap=cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p)
cb.set_label("\nFault energy ($mJ/m^2$)",fontsize=16)
ax.set_ylabel("$[0\,\overline{1}\,1]$ shift $\Delta$ (fraction)",fontsize=16)
ax.set_xlabel("$[2\,\overline{1}\,\overline{1}]$ shift $\Delta$ (fraction)",fontsize=16)
plt.show()
```
