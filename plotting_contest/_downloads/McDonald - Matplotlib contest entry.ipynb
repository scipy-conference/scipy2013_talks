```python
import numpy as np

def lorenz(t, y):
    """Compute Lorenz"""
    a = 16
    r = 45
    b = 4

    x_state = y[0]
    y_state = y[1]
    z_state = y[2]
    x_dt = a * (y_state - x_state)
    y_dt = (r*x_state) - y_state - (x_state * z_state)
    z_dt = (x_state * y_state) - (b * z_state)
    
    return np.array([x_dt, y_dt, z_dt])

def RK4(f, t0, y0, h, n):
    """Fourth order Runge-Kutta solver

    f:  some function
    t0: initial time point
    y0: nparray vector of variable states
    h:  time step
    n:  number of iterations

    returns a list of the variable states at each time point, and a list of
    timepoints
    """
    t_states = np.zeros(n+1, dtype=np.double)
    t_states[0] = t0
    y_states = np.zeros((n+1, y0.size), dtype=np.double)
    y_states[0] = y0
    
    t_current = t0

    for i in range(n):
        y_current = y_states[i]

        y_next = calc_step(f, h, t_current, y_current)
        t_next = t_current + h

        y_states[i+1] = y_next
        t_states[i+1] = t_next

        t_current = t_next

    return (t_states, y_states)

def calc_step(f, h, t, x):
    scaler = (1.0 / 6.0)
    t_half = h / 2
    
    k1 = h * f(t, x)
    k2 = h * f(t + t_half, x + k1 / 2)
    k3 = h * f(t + t_half, x + k2 / 2)
    k4 = h * f(t + h, x + k3)

    return x + (scaler * (k1 + (2 * k2) + (2 * k3) + k4))

def ARK4(f, t0, y0, h, n, eps):
    """Fourth order Runge-Kutta solver

    f:  some function
    t0: initial time point
    y0: nparray vector of variable states
    h:  time step
    n:  number of iterations

    returns a list of the variable states at each time point, and a list of
    timepoints
    """
    t_states = np.zeros(n+1, dtype=np.double)
    t_states[0] = t0
    y_states = np.zeros((n+1, y0.size), dtype=np.double)
    y_states[0] = y0

    t_current = t0
    
    for i in range(n):
        y_current = y_states[i]
        t_current = t_states[i]

        full_step = calc_step(f, h, t_current, y_current)
        half_step_1 = calc_step(f, h / 2, t_current, y_current)
        half_step_2 = calc_step(f, h / 2, t_current + (h / 2), half_step_1)

        norm = np.linalg.norm(full_step - half_step_2)
        h = h * ((eps / norm)**0.2)

        y_states[i+1] = half_step_2
        t_states[i+1] = t_current + h

    return (t_states, y_states)
```
